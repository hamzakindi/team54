# Import Required Libraries
Import the necessary libraries, including pandas and os.

```python
# Import Required Libraries
import pandas as pd
import os
```

# Set File Path
Set the file path to the folder containing the CSV files.

```python
# Set File Path
folder_path = r'C:\RnD\MedHack\MedHeck\archive\MedHack AI Hospital\csv'

# List all CSV files in the folder
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

# Display the list of CSV files
csv_files
```

# Load CSV Files
Use a loop to load all CSV files in the folder into separate pandas DataFrames.

```python
# Load CSV Files
dataframes = {}
for file in csv_files:
    file_path = os.path.join(folder_path, file)
    df_name = os.path.splitext(file)[0]
    dataframes[df_name] = pd.read_csv(file_path)

# Display the keys of the dataframes dictionary to confirm loading
dataframes.keys()
```

# Display DataFrames
Display the first few rows of each DataFrame to verify that the files have been loaded correctly.

```python
# Display DataFrames
for name, df in dataframes.items():
    print(f"DataFrame: {name}")
    display(df.head())
```
